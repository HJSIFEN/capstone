```python
keywords = ['축구', '음악', '야구', '달리기', '헬스', '아이폰', '노트북', '삼성전자', '옷', '신발']
```

```python
data_list, _ =get_data(keywords=keywords)
```

```python
import pandas as pd

df = pd.concat([pd.DataFrame(a) for a in data_list]).drop(['post_date'], axis = 1)
```

```python
df.to_csv('data.csv')
```

```python
import matplotlib.pyplot as plt


df.hist(figsize=(20, 20))
```

```python
df.info()
```

```python
df = df.dropna(axis = 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =42, test_size = 0.20)
```

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
print(standardScaler.fit(X_train))
X_train_standardScaled = standardScaler.transform(X_train)
```

```python
X_train_standardScaled.shape
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train_standardScaled, y_train)
```

```python
X_test_standardScaled = standardScaler.transform(X_test)
```

```python
plt.scatter(y_train, lr.predict(X_train_standardScaled), alpha=0.3)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, lr.predict(X_test_standardScaled), alpha = 0.3)
plt.xlim(0, 1000)
plt.ylim(0, 1000)
plt.show()
```

```python
from sklearn.svm import SVR

svr = SVR()
```

```python
svr.fit(X_train_standardScaled, y_train)
```

```python
plt.scatter(y_train, svr.predict(X_train_standardScaled), alpha = 0.3)
plt.xlim(0, 1000)
plt.ylim(0, 1000)
```

```python
import numpy as np
plt.scatter(y_test, svr.predict(X_test_standardScaled), alpha = 0.3)
plt.plot(np.arange(0,1000), np.arange(0, 1000))

plt.show()
```
